```python
x_train = train_df.iloc[:, :-1].values
y_train = train_df.iloc[:, 11:12].values

print(x_train.shape)
print(y_train.shape)

x_test = test_df.iloc[:, :].values
print(x_test.shape)
x_test = np.array(x_test)
print(x_test.shape)
```

```python
def temp_func(X):
    for i in range(0,len(X)):
       if X[i][2] == "3+":
           X[i][2] = 3
temp_func(x_train)
temp_func(x_test)   
train_df.info()
```

```python
from sklearn.impute import SimpleImputer

imputer1 = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer1 = imputer1.fit(x_train[:, 0:5])
x_train[:, 0:5] = imputer1.transform(x_train[:, 0:5])
x_test[:, 0:5] = imputer1.transform(x_test[:, 0:5])

imputer2 = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer2 = imputer2.fit(x_train[:, 8:10])
x_train[:, 8:10] = imputer2.transform(x_train[:, 8:10])
x_test[:, 8:10] = imputer2.transform(x_test[:, 8:10])

imputer3 = SimpleImputer(missing_values=np.nan, strategy='median')
imputer3= imputer3.fit(x_train[:, 7:8])
x_train[:, 7:8] = imputer3.transform(x_train[:, 7:8])
x_test[:, 7:8] = imputer3.transform(x_test[:, 7:8])
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder_x_0 = LabelEncoder()
x_train[:, 0] = labelencoder_x_0.fit_transform(x_train[:, 0])
x_test[:, 0] = labelencoder_x_0.transform(x_test[:, 0])

labelencoder_x_1 = LabelEncoder()
x_train[:, 1] = labelencoder_x_1.fit_transform(x_train[:, 1])
x_test[:, 1] = labelencoder_x_1.transform(x_test[:, 1])

labelencoder_x_2 = LabelEncoder()
x_train[:, 3] = labelencoder_x_2.fit_transform(x_train[:, 3])
x_test[:, 3] = labelencoder_x_2.transform(x_test[:, 3])

labelencoder_x_3 = LabelEncoder()
x_train[:, 4] = labelencoder_x_3.fit_transform(x_train[:, 4])
x_test[:, 4] = labelencoder_x_3.transform(x_test[:, 4])

labelencoder_x_4 = LabelEncoder()
x_train[:, 10] = labelencoder_x_4.fit_transform(x_train[:, 10])
x_test[:, 10] = labelencoder_x_4.transform(x_test[:, 10])


def dummy_drop(X):    
    c_10 = pd.DataFrame(X[:,10])
    c_10 = pd.get_dummies(c_10)
    c_10 = c_10.drop("0_0", axis = 1)
    
    c_2 = pd.DataFrame(X[:,2])
    c_2 = pd.get_dummies(X[:,2])
    c_2 = c_2.drop(0, axis = 1) 
    
    X = np.delete(X, [2, 10], axis = 1) 
    
    X = np.concatenate((X,c_2,c_10), axis = 1) 
    return X
x_train = dummy_drop(x_train)
x_test = dummy_drop(x_test)

```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
y_train=y_train.reshape(-1,1)
print(x_train.shape)
print(y_train.shape)

x_test = np.array(x_test)
print(x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train[:, 4:8] = sc_X.fit_transform(x_train[:, 4:8])
x_test[:, 4:8] = sc_X.transform(x_test[:, 4:8])
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy')
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_train)
cm = confusion_matrix(y_train, y_pred)
print (f"accuracy = {(cm[0][0]+cm[1][1])/(cm[0][0]+cm[1][1]+cm[0][1]+cm[1][0])}")
precission, recall, f_score, support = precision_recall_fscore_support(y_train, y_pred, pos_label = "Y", average = "binary")
print(f"precision = {precission}, recall = {recall}, f_score = {f_score}")#Different model evaluation can be used the good one is obtained through RandomForestClassifier
```
